```python
# Import necessary libraries
import pandas as pd
import seaborn as sns

# Load the Iris dataset
df = sns.load_dataset('iris')

# ---------------- Display Structure ----------------

#  Dataset shape (rows, columns)
print(" Shape of the dataset:")
print(df.shape)

#  Column names
print("\n Column names:")
print(df.columns.tolist())

# First 5 rows (sample data)
print("\n First 5 rows of the dataset:")
print(df.head())
```

**Code 5: Scatter Plot â€“ Petal Length vs Petal Width**


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Scatter plot: Petal Length vs Petal Width, colored by species
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='petal_length', y='petal_width', hue='species', palette='Set1')
plt.title('Scatter Plot: Petal Length vs Petal Width')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.grid(True)
plt.show()
```

This scatter plot shows the relationship between petal length and petal width, color-coded by flower species. It helps us visually separate species and detect patterns (e.g., Setosa has clearly smaller petals).

**Code 6: Histogram â€“ Sepal Length Distribution**


```python
# Histogram for Sepal Length
plt.figure(figsize=(8, 6))
sns.histplot(df['sepal_length'], kde=True, bins=15, color='orange')
plt.title('Histogram: Sepal Length Distribution')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

The histogram displays the frequency distribution of sepal_length values. The KDE curve shows the smooth distribution shape. This helps us see whether data is skewed, normal, or has outliers.

**Code 7: Box Plot â€“ Sepal Width by Species**


```python
# Corrected box plot for Sepal Width by Species (no FutureWarning)
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='species', y='sepal_width', hue='species', palette='Set2', dodge=False)
plt.title('Box Plot: Sepal Width by Species')
plt.xlabel('Species')
plt.ylabel('Sepal Width (cm)')
plt.legend([],[], frameon=False)  # Hides the redundant legend
plt.grid(True)
plt.show()

```

This box plot shows the distribution and spread of sepal_width for each species. It helps detect:

Outliers (dots outside whiskers)

Median (horizontal line inside the box)

Spread/variance of values



```python
# Import necessary visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset again just in case
import pandas as pd
df = sns.load_dataset('iris')

# ------------------ Scatter Plot (Seaborn + Matplotlib) ------------------
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='petal_length', y='petal_width', hue='species')
plt.title('Scatter Plot: Petal Length vs Petal Width')  # matplotlib title
plt.xlabel('Petal Length (cm)')                         # matplotlib xlabel
plt.ylabel('Petal Width (cm)')                          # matplotlib ylabel
plt.grid(True)
plt.show()

# ------------------ Histogram (Seaborn + Matplotlib) ------------------
plt.figure(figsize=(8, 5))
sns.histplot(df['sepal_length'], kde=True, bins=20, color='purple')
plt.title('Histogram: Sepal Length Distribution')  # matplotlib title
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# ------------------ Box Plot (Seaborn + Matplotlib) ------------------
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='species', y='sepal_width', hue='species', palette='Set2', dodge=False)
plt.title('Box Plot: Sepal Width by Species')  # matplotlib title
plt.xlabel('Species')
plt.ylabel('Sepal Width (cm)')
plt.legend([],[], frameon=False)  # Hide extra legend
plt.grid(True)
plt.show()
```
